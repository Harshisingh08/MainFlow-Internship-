```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
file_path = "/Users/harshitasingh/Downloads/tips.csv"
data = pd.read_csv(file_path)

# Display the first 10 rows of the DataFrame
display(data.head(10))

# Create a bar chart for 'day' vs 'tip'
plt.figure(figsize=(10, 6))
plt.bar(data['day'], data['tip'], color='blue')
plt.xlabel('Day')
plt.ylabel('Tip')
plt.title('Bar Chart of Tips by Day')
plt.xticks(rotation=45)
plt.legend(['Tip'])
plt.tight_layout()
plt.show()

# Create a line chart for 'day' vs 'tip'
plt.figure(figsize=(10, 6))
plt.plot(data['day'], data['tip'], marker='o', linestyle='-', color='green')
plt.xlabel('Day')
plt.ylabel('Tip')
plt.title('Line Chart of Tips by Day')
plt.xticks(rotation=45)
plt.legend(['Tip'])
plt.tight_layout()
plt.show()
```
