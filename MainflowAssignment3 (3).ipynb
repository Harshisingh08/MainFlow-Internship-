```python

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# Sample DataFrame
data = {
    'Month': ['January', 'February', 'March', 'April', 'May'],
    'Sales': [200, 150, 300, 400, 250],
    'Expenses': [100, 120, 180, 200, 150]
}
df = pd.DataFrame(data)

df.set_index('Month', inplace=True)

# bar chart
plt.figure(figsize=(10, 5))
plt.bar(df.index, df['Sales'], color='blue', label='Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales')
plt.legend()
plt.show()

# line chart 
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['Sales'], marker='o', linestyle='-', color='blue', label='Sales')
plt.plot(df.index, df['Expenses'], marker='o', linestyle='-', color='red', label='Expenses')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.title('Monthly Sales and Expenses')
plt.legend()
plt.show()
```
