```python
import pandas as pd
import matplotlib.pyplot as plt

# reading the database
data = pd.read_csv("/Users/harshitasingh/Downloads/tips.csv")
# Scatter plot with day against tip
plt.scatter(data['day'], data['tip'])
# Adding Title to the Plot
plt.title("Scatter Plot")
# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# reading the database
data = pd.read_csv("/Users/harshitasingh/Downloads/tips.csv")
# Bar chart with day against tip
plt.bar(data['day'], data['tip'])
plt.title("Bar Chart")
# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# reading the database
data = pd.read_csv("/Users/harshitasingh/Downloads/tips.csv")
# Line chart with tip and size columns
plt.plot(data['tip'], label='Tip')
plt.plot(data['size'], label='Size')
# Adding Title to the Plot
plt.title("Line Chart")
# Setting the X and Y labels
plt.xlabel('Index')
plt.ylabel('Value')
# Adding legend
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# reading the database
data = pd.read_csv("/Users/harshitasingh/Downloads/tips.csv")
# Histogram to visualize data distribution
plt.hist(data['tip'], bins=10, color='blue', edgecolor='black')
# Adding Title to the Plot
plt.title("Histogram")
# Setting the X and Y labels
plt.xlabel('Tip')
plt.ylabel('Frequency')
plt.show()
```
